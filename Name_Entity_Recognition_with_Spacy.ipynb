### <font color='#1dd1a1'> Packages

```python
# Importer les packages
import spacy
from spacy import displacy
!python -m spacy download en_core_web_lg
!python -m spacy download fr_core_news_lg
```

```python
# Extraire les packages en_core_web_lg et fr_core_news_lg
NER_english =spacy.load('en_core_web_lg')
NER_french =spacy.load('fr_core_news_lg')
```

### <font color='#1dd1a1'> Name Entity Recognition sur du texte en anglais

```python
# Texte
raw_text="""Legislative definitions of a federal agency are varied, and even contradictory.
The official United States Government Manual offers no definition.
While the Administrative Procedure Act definition of "agency" applies to most executive branch agencies,
Congress may define an agency however it chooses in enabling legislation, and through subsequent litigation
often involving the Freedom of Information Act and the Government in the Sunshine Act.
These further cloud attempts to enumerate a list of agencies.The executive branch of the federal government
includes the Executive Office of the President and the United States federal executive departments (whose secretaries belong to the Cabinet).
Employees of the majority of these agencies are considered civil servants.
The majority of the independent agencies of the United States government are also classified as executive agencies
(they are independent in that they are not subordinated under a Cabinet position).
There are a small number of independent agencies that are not considered part of the executive branch,
such as the Congressional Research Service and the United States Sentencing Commission, which are legislative and judicial agencies, respectively."""
```

####  <font color='#708090'> Appliquer le NER

```python
# Appliquer NER
text= NER_english(raw_text)
# Afficher les entits des noms
for word in text.ents:
  print(word.text,'------',word.label_)
```

```python
# ORG explication
spacy.explain('ORG')
```

```python
# GPE explication
spacy.explain('GPE')
```

```python
# LAW explication
spacy.explain('LAW')
```

##### <font color='#708090'> Visualiser les entitts de noms

```python
# Visualiser le NE's
displacy.render(text,style="ent",jupyter=True)
```
